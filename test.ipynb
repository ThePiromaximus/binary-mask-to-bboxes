```python
from binary_mask_to_bboxes import get_bounding_boxes
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
```

```python
# Open the image in ./imgs/mask.png
mask = np.array(Image.open('./imgs/mask.png'))/65535
bounding_boxes = get_bounding_boxes(mask)

# Print the bounding boxes
for box in bounding_boxes:
    print(box)
```

```python
# Plot the mask and the bounding boxes on top of it using matplotlib
plt.figure(figsize=(10,10))
plt.imshow(mask, cmap='gray')
ax = plt.gca()
for box in bounding_boxes:
    ax.add_patch(Rectangle((box[0], box[1]), box[2]-box[0], box[3]-box[1], linewidth=1, edgecolor='r', facecolor='none'))
plt.show()
```
